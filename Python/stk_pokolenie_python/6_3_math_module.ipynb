```python
import math
a = float(input())
b = float(input())
a1 = (a + b)/2
a2 = math.sqrt(a * b)
a3 = (2 * a * b)/(a + b)
a4 = math.sqrt((a**2 + b**2)/2)
print(a1)
print(a2)
print(a3)
print(a4)
```

```python
# —Ä–µ—à–µ–Ω–∏–µ —É—á–∏—Ç–µ–ª—è
import math

a = float(input())
b = float(input())

arithmetic_mean = (a + b) / 2

geometric_mean = math.sqrt(a * b)

harmonic_mean = 2 / ((1 / a) + (1 / b))

quadratic_mean = math.sqrt((a ** 2 + b ** 2) / 2)

print(arithmetic_mean)
print(geometric_mean)
print(harmonic_mean)
print(quadratic_mean)
```
